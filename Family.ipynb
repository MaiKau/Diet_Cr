```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import itertools

driver = webdriver.Chrome()                                                 
driver.implicitly_wait(5)
driver.get('https://foodsafety.family.com.tw/Web_FFD/Page/FFD_1_1.aspx#header')
driver.find_element_by_css_selector("#ctl00_ContentPlaceHolder1_btnPRODUCTID").click()   #Submit
time.sleep(5)

url_list=[]
for url in driver.find_elements_by_xpath("//a[@href]"):
    url_list.append(url.get_attribute('href'))
    for url_remove in url_list:
        if '.jpg' not in url_remove:url_list.remove(url_remove)
            
print(url_list)

driver.quit()
```
