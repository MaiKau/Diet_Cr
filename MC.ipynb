```python
from selenium import webdriver
from bs4 import BeautifulSoup,NavigableString,Tag
import requests
import pyodbc
import time
#-------------------------------------------------------各類別的網址--------------------------------------------------#
res=requests.get("https://www.mcdonalds.com/tw/zh-tw/full-menu.html",headers={'Connect':'close'},timeout=5)
soup=BeautifulSoup(res.text)
category_links=[]
for k in soup.find('ul',{'class':'mcd-category-page__links-section mcd-category-page__left-nav--sub-category'}).find_all('a'):
    try:
        category_links.append("http://www.mcdonalds.com"+k['href'])
    except:
        pass
print(category_links)
# -------------------------------------------------------各類別下的產品網址---------------------------------------------#
i=-1
product_link=[]
for h in category_links:
    i=i+1
    if i>10: #共9類category(>10全爬)
        break
    else:
        res2=requests.get(category_links[i],headers={'Connect':'close'},timeout=5)
        soup2=BeautifulSoup(res2.text)        
        for g in soup2.find('ul',{'class':'product-listing-list'}).find_all('a'):
            try:
                product_link.append("http://www.mcdonalds.com"+g['href'])
            except:
                pass
print(product_link)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import itertools

i=-1
for j in product_link:
    i=i+1
    if i>134: #共133個 Resource表格(>134全爬)
        break
    else:
        driver = webdriver.Chrome()                                                    #開啟chrome模擬器
        driver.implicitly_wait(5)                                                      #time out
        driver.get(product_link[i])
        driver.find_element_by_xpath("//*[@id='headingOne']/h3/a").click()
        time.sleep(5)
        table_list=[]
        title_list=[]
        num_list=['1','2','3','4','5','6','7','8','9']
        res=requests.get(product_link[i],headers={'Connect':'close'},timeout=5)
        soup=BeautifulSoup(res.text)
        time.sleep(3)
        for k in soup.find('div',{'class':'d-block d-md-none product-detail__main-content-mob-head'}).find_all('span'):
            time.sleep(1)
            title_list.append(k.text)
        j=0
        for g in num_list:
            j=j+1
            if j > 8:
                break
            else:
                si=str("tr["+num_list[j]+"]")
                for table in driver.find_elements_by_xpath("//*[@id='collapseOne']/div/div/div/div[1]/table/tbody/"+si+"/td[1]"):
                    time.sleep(1)
                    table_list.append(table.text)
                    for table in driver.find_elements_by_xpath("//*[@id='collapseOne']/div/div/div/div[1]/table/tbody/"+si+"/td[2]/span[1]"):
                        time.sleep(1)
                        table_list.append(table.text)
                
        table_list.insert(0,title_list[1])
        table_list.insert(10,'-')
        table_list.insert(14,'-')
        time.sleep(1)
        print(table_list)
        driver.quit()
```
