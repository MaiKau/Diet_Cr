```python
from selenium import webdriver
from bs4 import BeautifulSoup,NavigableString,Tag
import requests
import pyodbc
import time
#-------------------------------------------------------各類別的網址--------------------------------------------------#

category_links=['https://www.mos.com.tw/menu/set.aspx','https://www.mos.com.tw/menu/breakfast.aspx','https://www.mos.com.tw/menu/sideDishes.aspx','https://www.mos.com.tw/menu/soup.aspx','https://www.mos.com.tw/menu/beverage.aspx','https://www.mos.com.tw/menu/dessert.aspx']

#-------------------------------------------------------各產品的網址--------------------------------------------------#
product_links=[]
product_links2=[]
i=-1
for h in category_links:
    i=i+1
    if i>7: #共6類category(>7全爬)
        break
    else:
        res=requests.get(category_links[i],headers={'Connect':'close'},timeout=5)
        soup=BeautifulSoup(res.text)
        
        for k in soup.find('ul',{'class':'productsList'}).find_all('a'):
            try:
                product_links.append("http://www.mos.com.tw/menu/"+k['href'])
            except:
                pass
for x in product_links:                  #部分含#的連結為不要的,故排除
    if "#" not in x:
        product_links2.append(x)
                
print(product_links2)
```

```python
i=-1
for h in product_links2:
    i=i+1
    if i>118: #共117個產品(>118全爬)
        break
    else:
        res=requests.get(product_links2[i],headers={'Connect':'close'},timeout=5)
        soup=BeautifulSoup(res.text)
        table_list=[]
        for title in soup.find('h1'):
            table_list.append(title.text)    
        for k in soup.find('div',{'class':'facts'}).find_all('tr'):
            for td in k.find_all('td'):
                table_list.append(td.text)
        print(table_list)
```
