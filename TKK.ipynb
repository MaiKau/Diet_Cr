```python
from selenium import webdriver
from bs4 import BeautifulSoup,NavigableString,Tag
import requests
import pyodbc
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import itertools
driver = webdriver.Chrome()                                                    #開啟chrome模擬器
driver.implicitly_wait(5)                                                      #time out
driver.get('https://www.tkkinc.com.tw/sideDish.html')
driver.find_element_by_xpath("/html/body/div[2]/div[1]/div[4]/div/div[2]/section[1]/div/div[2]/div[4]/div[14]/a").click() #連三more展開
time.sleep(3)
driver.find_element_by_xpath("/html/body/div[2]/div[1]/div[4]/div/div[2]/section[1]/div/div[2]/div[4]/div[26]/a").click()
time.sleep(3)
driver.find_element_by_xpath("/html/body/div[2]/div[1]/div[4]/div/div[2]/section[1]/div/div[2]/div[4]/div[38]/a").click()
time.sleep(3)
link_list=[]
link_list2=[]
for link in driver.find_elements_by_xpath("//a[@href]"):
    link_list.append(link.get_attribute('href'))
for x in link_list:                  #含有sideDishProduct的連結為產品的,其於排除
    if "sideDishProduct" in x:
        link_list2.append(x)
driver.close()
print(link_list2)
```
